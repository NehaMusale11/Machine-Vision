```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('Screenshot (1).png')  # Changed to the uploaded image name
print("Image Loaded Successfully")
```

```python
if img is not None:
    cv2_imshow(img)
else:
    print("Error: 'img' is None. Check if the image file was loaded correctly (e.g., correct filename in cv2.imread()).")
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
print("Shape:", img.shape)
print("Height:", img.shape[0])
print("Width:", img.shape[1])
print("Channels:", img.shape[2])
print("Datatype:", img.dtype)
```

```python
blue = img[:,:,0]
green = img[:,:,1]
red = img[:,:,2]

cv2_imshow(blue)
cv2_imshow(green)
cv2_imshow(red)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)
```

```python
edges = cv2.Canny(gray, 100, 200)

cv2_imshow(edges)
```
